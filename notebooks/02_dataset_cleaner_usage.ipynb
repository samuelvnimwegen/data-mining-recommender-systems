# DatasetCleaner Demo

This notebook runs the cleaning pipeline and shows the profile report.

```python
from pathlib import Path
from src.configs.config import CleanerConfig
from src.dataloader import DatasetCleaner
```

```python
project_root_path = Path.cwd()
if not (project_root_path / "src").exists():
    project_root_path = project_root_path.parent
cleaner_config = CleanerConfig(
    movies_csv_path=project_root_path / "data" / "raw" / "movies.csv",
    ratings_csv_path=project_root_path / "data" / "raw" / "ratings_train.csv",
    output_directory_path=project_root_path / "data" / "processed" / "notebook_demo",
    enable_tfidf_features=False,
    enable_time_decay=True,
    time_decay_half_life_days=365.0,
    enable_strict_profile=False,
)
dataset_cleaner = DatasetCleaner(cleaner_config=cleaner_config)
movies_dataframe, ratings_dataframe, profile_report = dataset_cleaner.clean_datasets_with_report()
```

```python
print(f"Movies rows: {len(movies_dataframe)}")
print(f"Ratings rows: {len(ratings_dataframe)}")
print(f"Orphans removed: {profile_report.orphan_ratings_removed}")
print(f"Duplicates removed: {profile_report.duplicate_ratings_removed}")
print(f"Orphan ratio: {profile_report.orphan_ratio():.4f}")
print(f"Duplicate ratio: {profile_report.duplicate_ratio():.4f}")
```

```python
movies_output_path, ratings_output_path, saved_report = dataset_cleaner.clean_and_save_with_report()
print(f"Saved movies: {movies_output_path}")
print(f"Saved ratings: {ratings_output_path}")
saved_report
```
