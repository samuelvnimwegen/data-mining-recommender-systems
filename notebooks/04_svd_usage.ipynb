# SVD Demo

This notebook demonstrates the `SVDModel` wrapper and shows fit, predict and recommend examples.

```
import pandas as pd

# Import the SVD wrapper
from src.models.svd_model import SVDModel
```

```
# Build a small ratings dataframe (same shape as tests)
ratings_dataframe = pd.DataFrame(
    {
        "userId": [1, 1, 1, 2, 2, 3, 3, 3, 4, 4],
        "movieId": [1, 2, 3, 1, 4, 2, 3, 5, 1, 5],
        "rating": [4.0, 3.5, 2.0, 4.5, 3.0, 4.0, 3.0, 5.0, 2.5, 4.5],
    }
)
ratings_dataframe.head()
```

```
# Fit an SVD model and make a sample prediction and recommendations
model = SVDModel(number_of_factors=10, number_of_epochs=10, random_seed=7)
model.fit(ratings_dataframe)
# Predict rating for user 2 on movie 2
pred = model.predict_rating(2, 2)
print("Predicted rating for user 2 on movie 2:", pred)

# Top-3 recommendations for user 3 (unseen items only)
recs = model.recommend_top_n(3, number_of_recommendations=3)
print("Top-3 recommendations for user 3:")
for movie_id, score in recs:
    print(" ", movie_id, score)
```
