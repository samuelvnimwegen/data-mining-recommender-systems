# CleanerConfig Demo

This notebook shows how to create and validate `CleanerConfig`, and how to connect it to `DatasetCleaner`.

```python
from pathlib import Path
from pprint import pprint

from src.configs.config import CleanerConfig
from src.dataloader import DatasetCleaner
```

```python
# Resolve repository root when notebook starts in a subfolder.
project_root_path = Path.cwd()
if not (project_root_path / "src").exists():
    project_root_path = project_root_path.parent

print("Repository root:", project_root_path)
```

```python
# Create a default CleanerConfig using the project's raw data paths
cleaner_config = CleanerConfig(
    movies_csv_path=project_root_path / "data" / "raw" / "movies.csv",
    ratings_csv_path=project_root_path / "data" / "raw" / "ratings_train.csv",
    output_directory_path=project_root_path / "data" / "processed",
    enable_tfidf_features=False,
    enable_time_decay=True,
    time_decay_half_life_days=365.0,
    enable_strict_profile=False,
    max_orphan_ratio=0.05,
    max_duplicate_ratio=0.20,
)

pprint(cleaner_config)
```

```python
# This demonstrates validation in CleanerConfig.__post_init__
try:
    CleanerConfig(
        movies_csv_path=project_root_path / "data" / "raw" / "movies.csv",
        ratings_csv_path=project_root_path / "data" / "raw" / "ratings_train.csv",
        output_directory_path=project_root_path / "data" / "processed",
        time_decay_half_life_days=0.0,  # invalid -> should raise
    )
except ValueError as err:
    print("Validation caught as expected:", err)
```

```python
# Create the cleaner instance but do not run heavy cleaning unless requested
dataset_cleaner = DatasetCleaner(cleaner_config=cleaner_config)
print("DatasetCleaner created. Input movies file:", cleaner_config.movies_csv_path)

# Guarded run flag
run_cleaning = False
if run_cleaning:
    movies_df, ratings_df, report = dataset_cleaner.clean_datasets_with_report()
    print("Cleaning finished. Profile:")
    print(report)
else:
    print("Cleaning skipped. Set run_cleaning = True to run it.")
```
