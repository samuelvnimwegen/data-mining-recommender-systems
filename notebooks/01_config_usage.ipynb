# CleanerConfig Demo

This notebook shows how to create and validate , and how to connect it to .

```python
from pathlib import Path
from pprint import pprint

from src.configs.config import CleanerConfig
from src.dataloader import DatasetCleaner
```

```python
# Resolve repository root when notebook starts in a subfolder.
project_root_path = Path.cwd()
if not (project_root_path / "src").exists():
    project_root_path = project_root_path.parent

print("Repository root:", project_root_path)
```

```python
# Build a valid config used by the cleaner.
cleaner_config = CleanerConfig(
    movies_csv_path=project_root_path / "data" / "raw" / "movies.csv",
    ratings_csv_path=project_root_path / "data" / "raw" / "ratings_train.csv",
    output_directory_path=project_root_path / "data" / "processed",
    enable_tfidf_features=False,
    enable_time_decay=True,
    time_decay_half_life_days=365.0,
    enable_strict_profile=False,
    max_orphan_ratio=0.05,
    max_duplicate_ratio=0.20,
)

pprint(cleaner_config.__dict__)
```

```python
# Show validation behavior for invalid values.
try:
    CleanerConfig(
        movies_csv_path=project_root_path / "data" / "raw" / "movies.csv",
        ratings_csv_path=project_root_path / "data" / "raw" / "ratings_train.csv",
        output_directory_path=project_root_path / "data" / "processed",
        max_orphan_ratio=2.0,
    )
except ValueError as error:
    print("Validation works:", error)
```

```python
# Create the cleaner object.
dataset_cleaner = DatasetCleaner(cleaner_config=cleaner_config)
print("Cleaner created successfully.")
```

```python
# Keep the run guarded for quick notebook startup.
run_cleaning = False

if run_cleaning:
    cleaned_movies_dataframe, cleaned_ratings_dataframe, profile_report = dataset_cleaner.clean_datasets_with_report()
    print("Movies rows:", len(cleaned_movies_dataframe))
    print("Ratings rows:", len(cleaned_ratings_dataframe))
    print("Profile report:", profile_report)
else:
    print("Cleaning skipped. Set run_cleaning = True to run the pipeline.")
```
