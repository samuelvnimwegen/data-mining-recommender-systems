# ItemKNN Demo

This notebook demonstrates the Item-based KNN wrapper (`ItemKNNModel`) and shows fit, predict and recommend examples.

```
import pandas as pd

# Import the ItemKNN wrapper
from src.models.item_knn_model import ItemKNNModel
```

```
# Build a small ratings dataframe (same shape as tests)
ratings_dataframe = pd.DataFrame(
    {
        "userId": [1, 1, 1, 2, 2, 3, 3, 3, 4, 4],
        "movieId": [1, 2, 3, 1, 4, 2, 3, 5, 1, 5],
        "rating": [4.0, 3.5, 2.0, 4.5, 3.0, 4.0, 3.0, 5.0, 2.5, 4.5],
    }
)
ratings_dataframe.head()
```

```
# Fit an ItemKNN model and make a sample prediction and recommendations
model = ItemKNNModel(number_of_neighbors=2, minimum_neighbors=1)
model.fit(ratings_dataframe)
# Predict rating for user 1 on movie 4 (user hasn't rated movie 4 in this small set)
pred = model.predict_rating(1, 4)
print("Predicted rating for user 1 on movie 4:", pred)

# Top-3 recommendations for user 1 (unseen items only)
recs = model.recommend_top_n(1, number_of_recommendations=3)
print("Top-3 recommendations for user 1:")
for movie_id, score in recs:
    print(" ", movie_id, score)
```
